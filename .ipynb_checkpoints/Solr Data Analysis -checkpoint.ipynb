```python
import pandas as pd
import numpy as np
```

```python
clientsToFbBrands = pd.read_csv("data/clientToFbBrands.csv")
clientsToIgBrands = pd.read_csv("data/clientToIgBrands.csv")

joined_df = clientsToFbBrands.merge(clientsToIgBrands, on="client_id", how="outer")
joined_df.head(10)
joined_df = joined_df.replace(np.nan, ' ', regex=True)
joined_df['facebookbrands'] = joined_df['facebookbrands'].str.slice(1, -1)
joined_df['instagrambrands'] = joined_df['instagrambrands'].str.slice(1, -1)
joined_df['combined'] = joined_df['instagrambrands'].str.cat(joined_df['facebookbrands'], sep=",")

formatted_joined_df = joined_df.drop(['instagrambrands', 'facebookbrands'], axis=1)
```

```python
from collections import defaultdict
query_id_to_client_id = defaultdict(str)

def query_ids_to_map(row):
    query_ids = row['combined'].split(',')
    for query_id in query_ids:
        query_id_to_client_id[query_id] = row['client_id']

x = formatted_joined_df.apply(query_ids_to_map, axis=1)

query_ids = [key for key in query_id_to_client_id]
print(query_ids)
```

```python
import json

client_query_count = []
with open('data/testresults_json.js') as json_file:
    x = json.load(json_file)
    facebook_buckets = x['facebook_posts']['buckets'][0]['query_id']['buckets']
    for query_dict in facebook_buckets:
        client_query_count.append((query_dict['count'], query_id_to_client_id[str(query_dict['val'])]))
    count = 0
    keys = {}
    for _, id in client_query_count:
        if id and id not in keys:
            count += 1
        keys[id] = 0
    print(count)
    
        
        
    
```
